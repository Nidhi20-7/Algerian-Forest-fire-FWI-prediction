## Algerian fire dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
df.head()
```

```python
df.info()
```

## Data cleaning

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
data=df
```

```python
df.info()
```

```python
data[['Region']]=data[['Region']].astype(int)
```

```python
data.head()
```

```python
data=data.dropna().reset_index(drop=True)
```

```python
data.isnull().sum()
```

```python
data.iloc[[122]]
```

```python
data=data.drop(122).reset_index(drop=True)
```

```python
data.columns
```

```python
data.columns=data.columns.str.strip()
data.columns
```

```python
cat_col=[]
for col in data.columns:
    if data[col].dtype=='object':
        cat_col.append(col)
```

```python
data[['day', 'month', 'year', 'Temperature', 'RH','Ws']]=data[['day', 'month', 'year', 'Temperature', 'RH','Ws']].astype(int)
```

```python
data.head()
```

```python
obj=[feat for feat in data.columns if data[feat].dtypes=='object']
```

```python
for i in obj :
    if i!='Classes':
        data[i]=data[i].astype(float)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.to_csv('Cleaned_algerian_data',index=False)
```

## EDA

```python
df_copy=data.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
df_copy.columns=df_copy.columns.str.strip()
```

```python
df_copy['Classes'].value_counts()
```

```python
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
perct=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
lbl=['Fire','Not fire']
plt.figure(figsize=(5,3))
plt.pie(perct,labels=lbl,autopct='%1.1f%%')
```

## Correlation

```python
df_copy.corr()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df_copy.corr(),annot=True)
```

```python
data['Classes']=np.where(data['Classes'].str.contains('not fire'),'not fire','fire')
```

```python
dftemp=data.loc[data['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',data=data,hue='Classes')
plt.ylabel('Number of fires')
plt.xlabel('Month')
plt.title("No. of fires of Region 1") 
```

```python
dftemp=data.loc[data['Region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',data=data,hue='Classes')
plt.ylabel('Number of fires')
plt.xlabel('Month')
plt.title("No. of fires of Region 0") 
```
